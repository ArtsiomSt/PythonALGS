```python
import numpy as np
import matplotlib.pyplot as plt
import random
from sklearn.linear_model import LinearRegression
import random
```

```python
x = np.random.uniform(0, 40, 200)
x = np.sort(x)
y = -3*x + 20 + np.random.normal(0, 15, len(x))
```

```python
def update_w_b(alpha, w, b, x, y, start):
    dx_dw = 0
    dx_db = 0
    for i, value in enumerate(x[start:start+50]):
        dx_dw += -2*(y[i]-(w*x[i]+b))*x[i]
        dx_db += -2*(y[i]-(w*x[i]+b))
    w = w - alpha*(1/float(len(x)))*dx_dw
    b = b - alpha*(1/float(len(x)))*dx_db
    return w, b
```

```python
def get_loss(x, y, w, b):
    loss = 0
    for i in range(len(x)):
        loss += (y[i] - (w*x[i]+b))**2
    return loss*float(1/len(x))
```

```python
def train(x, y, w, b, alpha, epochs):
    start = random.randrange(0, len(x)-100)
    for epoch in range(epochs):
        w, b = update_w_b(alpha , w, b, x, y, start)
        loss = get_loss(x, y, w, b)
        print(f':epoch: {epoch} -- loss: {loss}')
    return w, b

w, b = train(x, y, 0, 0, 0.1, 300)
print(w, b)
```

```python
def predict(x, w, b):
    p = lambda x: w*x+b
    return np.array(list(map(p, x)))
```

```python
yy = predict(x, w, b)
print(w, b)
```

```python
model = LinearRegression()
model.fit(x.reshape(-1,1),y.reshape(-1,1))
plt.plot(x,y, 'bo')
plt.plot(x,yy, 'r-')
plt.plot(x, model.predict(x.reshape(-1, 1)), 'g-')
plt.show()
```
