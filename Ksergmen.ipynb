```python
import numpy as np
import math
import random
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(0, 20, 200)
x = np.sort(x)
```

```python
y = 1*x - 25 + np.random.normal(0, 10, len(x))
```

```python
x2 = np.random.uniform(0, 20, 200)
x2 = np.sort(x)
```

```python
y2 = 0.5*x2 - 50 + np.random.normal(0, 10, len(x))
```

```python
plt.plot(x, y, 'go')
plt.plot(x2, y2, 'bo')
```

```python
dots_x = np.hstack([x, x2])
dots_y = np.hstack([y, y2])
```

```python
plt.plot(dots_x, dots_y, 'go')
```

```python
def categories(x, y):
    centeroid_one = random.randrange(0, len(x))
    centeroid_two = random.randrange(0, len(x))  
    class_one = []
    class_two = []
    epochs = 1000
    for e in range(epochs):
        centroid_one_x = x[centeroid_one]
        centroid_one_y = y[centeroid_one]
        centroid_two_x = x[centeroid_two]
        centroid_two_y = y[centeroid_two]
        for i in range(len(x)):
            if math.dist((x[i], y[i]), (centroid_one_x, centroid_one_y)) < math.dist((x[i], y[i]), (centroid_two_x, centroid_two_y)):
                class_one.append(i)
            else:
                class_two.append(i)
        centroid_one_x  = sum(x[class_one])/float(1/len(x[class_one]))
        centroid_one_y = sum(y[class_one])/float(1/len(x[class_one]))
        centroid_two_x = sum(x[class_two])/float(1/len(class_two))
        centroid_one_y = sum(y[class_two])/float(1/len(class_two))
        if e == epochs - 1:
            return class_one, class_two
        else:
            class_one.clear()
            class_two.clear()
```

```python
one, two = categories(dots_x, dots_y)
plt.plot(dots_x[one], dots_y[one], 'ro')
plt.plot(dots_x[two], dots_y[two], 'go')
plt.show()
```
